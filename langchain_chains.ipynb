- Sequential chain

```python
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv

load_dotenv()

llm = ChatOpenAI()

llm.invoke("Tell me about Machine learning ")
```

```python
# Sequential Chain
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv

load_dotenv()

prompt1=PromptTemplate(
    template='Generate a detailed Report on the {topic}',
    input_variables=['topic']
)

prompt2=PromptTemplate(
    template='Give a 5 points Bullet points on the {information}',
    input_variables=['information']
)

model = ChatGoogleGenerativeAI(model='gemini-2.0-flash')

parser = StrOutputParser()

chain = prompt1 | model | parser | prompt2 | model | parser

res = chain.invoke({"topic":"IT Employment Ratio in india From 2021 to 2025"})

print(res)
```

```python
# parallel Chain
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_openai.chat_models import ChatOpenAI
from langchain_core.prompts import PromptTemplate
from langchain_core.output_parsers import StrOutputParser
from dotenv import load_dotenv
from langchain.schema.runnable import RunnableParallel

model1 = ChatGoogleGenerativeAI(model="gemini-2.0-flash")
model2 = ChatOpenAI(model="gpt-3.5-turbo")

prompt1 = PromptTemplate(
    template="Generate a Short and Simple notes on the following text \n {text}",
    input_variables=['text']
)

prompt2 = PromptTemplate(
    template='Generate 5 short question and answer froms the following text \n {text}',
    input_variables=['text']
)

prompt3 = PromptTemplate(
    template='Marge the Provided notes and Quiz into a Single Document \n  notes --> {notes}\n Quiz --> {quiz}',
    input_variables=['notes','quiz']
)

parser = StrOutputParser()

parallel_chain = RunnableParallel({
    'notes': prompt1 | model1 | parser,
    'quiz' : prompt2 | model2 | parser
})

merge_chain = prompt3 | model1 | parser
chain = parallel_chain | merge_chain | parser

text = """ Supervised learning algorithms learn from labeled data, where the input data (features) and the corresponding output (labels) are provided. The goal is to make predictions based on input data.

Common Supervised Learning Algorithms:
Linear Regression

Purpose: Used for regression tasks where the goal is to predict a continuous value.

Working: It assumes a linear relationship between the input variables and the output. The model learns the coefficients of the linear equation that best fit the training data.

Example: Predicting house prices based on features like size, location, etc.

Logistic Regression

Purpose: Used for binary classification problems.

Working: It predicts the probability of a binary outcome (0 or 1) based on the input features. The output is transformed through the logistic function (sigmoid) to give a probability.

Example: Predicting whether a customer will buy a product or not.

Decision Trees

Purpose: Used for both classification and regression tasks.

Working: The tree splits the data into branches based on feature values, making decisions at each node. The tree is constructed by recursively splitting the dataset in a way that maximizes information gain.

Example: Classifying emails as spam or not.

Random Forest

Purpose: An ensemble learning method that combines multiple decision trees.

Working: It generates a collection of decision trees using bootstrapped samples of the data. Each tree is trained independently, and the final prediction is based on the majority vote (classification) or average (regression) from all trees.

Example: Predicting customer churn in telecom companies.

Support Vector Machine (SVM)

Purpose: Used for classification tasks, especially with high-dimensional data.

Working: SVM finds the hyperplane that best separates the classes in the feature space. The optimal hyperplane is the one with the maximum margin between the classes.

Example: Handwriting recognition.

K-Nearest Neighbors (KNN)

Purpose: Used for both classification and regression.

Working: It classifies or predicts the output based on the majority class or average of the K nearest points to the input data point.

Example: Predicting whether a person will be interested in a product based on similar users.

Naive Bayes

Purpose: Used for classification tasks, especially text classification.

Working: Based on Bayes’ Theorem, it assumes that the features are independent. It calculates the probability of each class given the features and assigns the class with the highest probability.

Example: Spam email classification

"""

res = chain.invoke({'text':text})
```
